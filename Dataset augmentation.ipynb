```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import load_img
import numpy as np
import argparse
import os
import cv2
```

```python
base_image_direc = './dataset/new_direc/images/'
base_mask_direc = './dataset/new_direc/masks/'

image_names = os.listdir(base_image_direc)
curr_id = 1052

for i in image_names:
    img = cv2.imread(base_image_direc + i)
    mask = cv2.imread(base_mask_direc + i)
    
    img_h = cv2.flip(img, 0)
    mask_h = cv2.flip(mask, 0)
    cv2.imwrite(base_image_direc + str(curr_id) + '.jpg', img_h)
    cv2.imwrite(base_mask_direc + str(curr_id) + '.jpg', mask_h)
    curr_id = curr_id + 1
    
    img_v = cv2.flip(img, 1)
    mask_v = cv2.flip(mask, 1)
    cv2.imwrite(base_image_direc + str(curr_id) + '.jpg', img_v)
    cv2.imwrite(base_mask_direc + str(curr_id) + '.jpg', mask_v)
    curr_id = curr_id + 1
    
    img_hv = cv2.flip(img_h, 1)
    mask_hv = cv2.flip(mask_h, 1)
    cv2.imwrite(base_image_direc + str(curr_id) + '.jpg', img_hv)
    cv2.imwrite(base_mask_direc + str(curr_id) + '.jpg', mask_hv)
    curr_id = curr_id + 1
```
