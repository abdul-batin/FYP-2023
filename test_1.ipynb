```python
# Python program to explain
# cv2.polylines() method
import cv2
import numpy as np
import copy

# path
path = 'dataset/Radiographs/315.jpg'

def get_polygon_vertices(image_id, my_dict):
    return np.array(my_dict[image_id]['Label']['objects'][0]['polygons'])
    

def buildImage(image_id):
    # Reading an image in default
    image = cv2.imread('dataset/Radiographs/' + image_id + '.jpg')

    # Window name in which image is
    # displayed
    window_name = 'Image'

    # Polygon corner points coordinates
    pts = get_polygon_vertices(image_id, my_dict)
    if pts.ndim != 0:
#     pts = np.array(expert_data[713]['Label']['objects'][0]['polygons'][0], np.int32)
        pts = pts.reshape((-1, 1, 2))

        isClosed = True
        # Green color in BGR
        color = (0, 255, 0)
        # Line thickness of 8 px
        thickness = 2
        # Using cv2.polylines() method
        # Draw a Green polygon with
        # thickness of 1 px
        image = cv2.polylines(image, [pts], isClosed, color,thickness)
    
    cv2.imshow('Image', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def buildRadioGraph(image_id):
    # Window name in which image is
    # displayed
    image = cv2.imread('dataset/Radiographs/' + image_id + '.jpg')
    window_name = 'Image'
    
    for tooth in my_dict['763']['Label']['objects']:
        if tooth['polygons'] != 'none':
            for tp in tooth['polygons']:
                pts = np.array(tp)
                pts = pts.reshape((-1, 1, 2))
                isClosed = True
                color = (0, 255, 0)
                thickness = 2
                image = cv2.polylines(image, [pts], isClosed, color, thickness)
        
    for tooth in data[4]['Label']['objects']:  
            xmin = tooth['bounding box'][0]
            ymin = tooth['bounding box'][1]
            xmax = tooth['bounding box'][2]
            ymax = tooth['bounding box'][3]

            start_point = (ymin, xmin)#(xmin, ymin)
            end_point = (ymax, xmax) #(xmax, ymax)

            color = (255, 0, 0)
            thickness = 2

            image = cv2.rectangle(image, start_point, end_point, color, thickness)
            cv2.putText(image, tooth['title'], (ymin, xmin-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36,255,12), 2)

    cv2.imshow(window_name, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
